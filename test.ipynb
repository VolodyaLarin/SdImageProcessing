```python
from transformers import DPTImageProcessor, DPTForDepthEstimation
import torch
import numpy as np
from PIL import Image
import requests

from transformers import pipeline

pipe = pipeline(task="depth-estimation", model="Intel/dpt-swinv2-tiny-256")

```

```python
def get_concat_h(im1, im2):
    dst = Image.new('RGB', (im1.width + im2.width, im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width, 0))
    return dst
```

```python
import cv2

video = 'video.mp4'
vidcap = cv2.VideoCapture(video)
count = 0

import time
import numpy

success, image = vidcap.read()
input = Image.fromarray(image)

# size = (128,128)
# input.thumbnail(size, Image.Resampling.LANCZOS)

print(input.size)

video_out=cv2.VideoWriter('/workspaces/SdImageProcessing/video_out.mp4', cv2.VideoWriter_fourcc(*"mp4v"), 20.0, (input.size[0]*2, input.size[1]))

timeline = []
while vidcap.isOpened():
    success, image = vidcap.read()
    if success:
        if count % 1 == 0:
            start = time.time()
             
            input = Image.fromarray(image)
            # input.thumbnail(size, Image.Resampling.LANCZOS)

            result = pipe(input)
            # display(result)
            depth = result["depth"]
            
            end = time.time() - start

            # display(get_concat_h(input, depth))

            video_out.write(numpy.array(get_concat_h(input, depth)))

            timeline.append(end)
        count += 1
    else:
        break
cv2.destroyAllWindows()
vidcap.release()
video_out.release()
display(timeline)

```
